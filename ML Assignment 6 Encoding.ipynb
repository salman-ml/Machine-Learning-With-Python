# Odinal Encoder

```python
df9.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df9.sex.unique()
```

```python
sex_new=['female', 'male']
```

```python
Ordinal=OrdinalEncoder(categories=[sex_new])
encoded=Ordinal.fit_transform(df9[['sex']])
```

```python
encoded
```

```python
new_sex_name=pd.DataFrame(encoded, columns=['sex'])
```

```python
new_sex_name.head()
```

```python
df9.sex=new_sex_name
```

```python
df9.head()
```

```python
df9.smoker.unique()
```

```python
df9.region.unique() 
```

```python
cols = ['smoker','region']
for col in cols:
    unique = df9[col].unique()
    df9[col] = OrdinalEncoder(categories=[unique]).fit_transform(df9[[col]])
```
